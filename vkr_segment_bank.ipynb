**3. Добавление новых признаков**


```python
# Добавлю новый признак "Возраст"
# Получаем текущую дату
today = pd.Timestamp.now()

# Разница лет
df["Возраст"] = today.year - df["Дата рождения"].dt.year

# Коррекция возраста, если оно не произошло еще
birthday = (
    (df["Дата рождения"].dt.month > today.month) |
    ((df["Дата рождения"].dt.month == today.month) &
     (df["Дата рождения"].dt.day > today.day))
)

df.loc[birthday, "Возраст"] -= 1

# Проверка результата
df[["Дата рождения", "Возраст"]].head()
```
